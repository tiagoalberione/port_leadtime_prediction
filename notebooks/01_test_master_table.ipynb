```python
# Project path

from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print(PROJECT_ROOT)
```

```python
# Imports

import pandas as pd

from src.data_sources.port_call import load_port_call_files, process_port_call
from src.processing.master_table import build_master_calls
```

```python
# Load and process data

RAW_PORT_CALL_DIR = PROJECT_ROOT / "data" / "raw" / "estadia"

df_raw = load_port_call_files(RAW_PORT_CALL_DIR)
df_port_call = process_port_call(df_raw)

print(df_port_call.shape)
df_port_call.head()
```

```python
# Checking for duplicates in port_call_id

print("Rows:", len(df_port_call))
print("Unique port_call_id:", df_port_call["port_call_id"].nunique())
```

```python
# Investigating duplicates by port_call_id

dup_counts = (
    df_port_call.groupby("port_call_id")
    .size()
    .reset_index(name="n_rows")
    .sort_values("n_rows", ascending=False)
)

dup_counts.head(20)
```

```python
# Building the master table

df_master = build_master_calls(df_port_call)

print(df_master.shape)
df_master.head()
```

```python
# Checking for duplicates in master table

print("Rows in master:", len(df_master))
print("Unique call_id in master:", df_master["port_call_id"].nunique())
print("Is unique:", len(df_master) == df_master["port_call_id"].nunique())
```

```python
# Checking main timestamps

date_cols = [
    "arrival_port_ts",
    "berthing_ts",
    "unberthing_ts",
    "departure_port_ts",
]

df_master[date_cols].head(10)
```

```python
# Coverage by column

df_master.isna().mean().sort_values(ascending=False)
```

```python
# Saving the result

output_path = PROJECT_ROOT / "data" / "interim" / "master_calls.parquet"
output_path.parent.mkdir(parents=True, exist_ok=True)

df_master.to_parquet(output_path, index=False)

print(f"Saved to: {output_path}")
```
