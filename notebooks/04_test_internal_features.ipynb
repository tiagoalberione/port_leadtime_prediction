```python
# Setup

from pathlib import Path
import sys

PROJECT_ROOT = Path.cwd().parent
if str(PROJECT_ROOT) not in sys.path:
    sys.path.append(str(PROJECT_ROOT))

print(PROJECT_ROOT)
```

```python
# Imports

import pandas as pd

from src.features.calendar import create_calendar_features
from src.features.congestion import create_basic_congestion_features
```

```python
# Load target base

target_base_path = PROJECT_ROOT / "data" / "interim" / "target_base.parquet"

df_target = pd.read_parquet(target_base_path)

print(df_target.shape)
df_target.head()
```

```python
# Create calendar features

df_feat = create_calendar_features(df_target, ref_col="arrival_port_ts")

print(df_feat.shape)
df_feat.head()
```

```python
# Inspect the new calendar columns

calendar_cols = [
    "arrival_year",
    "arrival_month",
    "arrival_quarter",
    "arrival_weekofyear",
    "arrival_day",
    "arrival_dayofweek",
    "arrival_hour",
    "arrival_is_weekend",
    "arrival_shift",
    "arrival_season",
]

df_feat[calendar_cols].head(10)
```

```python
# Creating congestion features

df_feat = create_basic_congestion_features(df_feat)

print(df_feat.shape)
df_feat.head()
```

```python
# Inspecting the new congestion columns

congestion_cols = [
    "avg_wait_prev_20_calls_port",
    "avg_operation_prev_20_calls_port",
    "std_wait_prev_20_calls_port",
    "arrival_date",
    "arrivals_same_day_port",
    "arrivals_prev_day_port",
    "arrivals_prev_7d_avg_port",
]

df_feat[congestion_cols].head(20)
```

```python
# Checking for missing values in the new features

new_feature_cols = calendar_cols + congestion_cols

df_feat[new_feature_cols].isna().mean().sort_values(ascending=False)
```

```python
# Saving results

interim_dir = PROJECT_ROOT / "data" / "interim"
interim_dir.mkdir(parents=True, exist_ok=True)

internal_features_path = interim_dir / "internal_features.parquet"

df_feat.to_parquet(internal_features_path, index=False)

print(f"Internal features base saved to: {internal_features_path}")
```
